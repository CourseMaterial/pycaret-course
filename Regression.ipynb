```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
from pycaret.datasets import get_data
from pycaret.regression import *
mpl.rcParams['figure.dpi'] = 300

data = get_data('insurance')
```

```python
data.info()
```

```python
numeric = ['bmi', 'age', 'charges']

data[numeric].hist(bins=20, figsize = (12,6))
plt.show()
```

```python
categorical = ['sex', 'children', 'smoker', 'region']
color = ['C0', 'C1', 'C2', 'C3', 'C4', 'C5']

fig, axes = plt.subplots(2, 2, figsize = (16,14))

for ax, col in zip(axes.flatten(), categorical) :
    data[col].value_counts(normalize = 'index').plot(kind='bar', ax=ax, color = color)
    ax.set_xlabel(col)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(18,10))
for ax, col in zip(axes.flatten(), categorical):
    sns.histplot(data, x='charges', hue=col,
               multiple='stack', ax=ax)
```
